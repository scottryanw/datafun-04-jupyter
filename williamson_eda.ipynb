```python
## Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

```python
## Initial data inspection
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

```python
## Print Descriptive Statistics
print(df.describe())
```

```python
# Inspect histogram by numerical column
df['sepal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

- Sepal width has a fairly normal distribution.
- Sepal length is close to normal distribution but has a slightly broader distribution of lengths.
- Petal length and petal width have a bimodal distribution with a heavy bias toward low values. 

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show distribution of Petal Length by Species
sns.catplot(data=df, x='species', y='petal_length', kind='box')

# Show all plots
plt.show()
```

- Each species has 50 datapoints.
- Each species have consistent distributions of petal length, especially setosa.

```python
# Renaming columns
df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)
df.rename(columns={'sepal_width': 'Sepal Width'}, inplace=True)
df.rename(columns={'petal_length': 'Petal Length'}, inplace=True)
df.rename(columns={'petal_width': 'Petal Width'}, inplace=True)

# Adding a new column
df['Sepal Area'] = df['Sepal Length'] * df['Sepal Width']

# Add new columns
```

```python
# Initial Visualizations
sns.pairplot(df, hue='species')
# Other Visualizations
plt.show()
```

- There appears to be a strong correlation in all species between petal width and petal length. 
- The setosa species has a very consistant petal length and petal width with almost no variation for different sepal length and widths. 
- There is a reasonably strong correlation between petal length and sepal length for the versicolor and virginica species.
